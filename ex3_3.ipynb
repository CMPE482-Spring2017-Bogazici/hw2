Vector and matrix $p$-norms are related by various inequalities, often involving the dimensions $m$ or $n$. For each of the following, verify the inequality and give an example of a nonzero vector or matrix (for general $m$,$n$) for which equality is achieved. In this problem $x$ is an $m$-vector and $A$ is an $m \times n$ matrix.

### (a) $\left\lVert x\right\rVert_{\infty} \leq \left\lVert x\right\rVert_2$

Given the definitions

$$\begin{align}
\left\lVert x\right\rVert_{\infty} &= max\{\lvert x_1\rvert, \lvert x_2\rvert, \dots, \lvert x_m\rvert\} \\
\left\lVert x\right\rVert_2 &= (x_1^2 + x_2^2 + \dots + x_m^2)^{\frac{1}{2}}
\end{align}$$

We can write

$$\begin{align}
\left\lVert x\right\rVert_{\infty} &= \lvert x_i\rvert \qquad\textit{for some } i \in 1, 2, \dots, m \\
\left\lVert x\right\rVert_{\infty}^2 &= x_i^2
\end{align}$$

We modify our original inequality as follows

$$\begin{align}
\left\lVert x\right\rVert_{\infty} &\leq \left\lVert x\right\rVert_2 \\
\left\lVert x\right\rVert_{\infty}^2 &\leq \left\lVert x\right\rVert_2^2 \\
x_i^2 &\leq x_1^2 + x_2^2 + \dots + x_m^2 \\
0 &\leq x_1^2 + \dots + x_{i-1}^2 + x_{i+1}^2 + \dots + x_m^2
\end{align}$$

Since $\forall x, x^2 \geq 0$, the resulting inequality always holds.

**Equality condition**

For the vector $\begin{bmatrix}0 \\ x_j \\ \vdots \\ 0\end{bmatrix}, 1 \leq j \leq m \text{ and } x_j \in \mathbb{R}$, the given two norms are equal.

### (b) $\left\lVert x\right\rVert_2 \leq \sqrt{m} \left\lVert x\right\rVert_{\infty}$

In addition to the definitions of $2$ and $\infty$ norms above, we again write the $\infty$-norm as follows

$$\begin{align}
\left\lVert x\right\rVert_{\infty} &= \lvert x_i\rvert \qquad\textit{for some } i \in 1, 2, \dots, m \\
\left\lVert x\right\rVert_{\infty}^2 &= x_i^2
\end{align}$$

We modify our original inequality as follows

$$\begin{align}
\left\lVert x\right\rVert_2 &\leq \sqrt{m} \left\lVert x\right\rVert_{\infty} \\
\left\lVert x\right\rVert_2^2 &\leq m\left\lVert x\right\rVert_{\infty}^2 \\
x_1^2 + x_2^2 + \dots + x_m^2 &\leq mx_i^2 \\
x_1^2 + x_2^2 + \dots + x_m^2 &\leq x_i^2 + x_i^2 + \dots + x_i^2 \\
0 &\leq (x_i^2 - x_1^2) + (x_i^2 - x_2^2) + \dots + (x_i^2 - x_m^2)
\end{align}$$

Since we know that $\forall j\in 1, 2, \dots, m \quad x_i^2 \geq x_j^2$ from the definition of $\infty$-norm, we conclude that the inequality always holds.

**Equality condition**

For the vector $\begin{bmatrix}x \\ x \\ \vdots \\ x\end{bmatrix}, x \in \mathbb{R}$, the given inequality becomes an equality.

### (c) $\left\lVert A\right\rVert_{\infty} \leq \sqrt{n}\left\lVert A\right\rVert_2$

From **(a)** we know that for a given $m \times 1$ vector $x$, 

$$\begin{align}
\left\lVert x\right\rVert_{\infty} \leq \left\lVert x\right\rVert_2
\end{align}$$

Let $y$ be any $n \times 1$ vector and $A$ an $m \times n$ matrix. Then $Ay$ is an $m \times 1$ vector and thus satisfies 

$$\begin{align}
\left\lVert Ay\right\rVert_{\infty} \leq \left\lVert Ay\right\rVert_2
\end{align}$$

If we take the supremum of both sides

$$\begin{align}
\textit{sup}_{\left\lVert y\right\rVert_n = 1} \left\lVert Ay\right\rVert_{\infty} \leq \textit{sup}_{\left\lVert y\right\rVert_n = 1} \left\lVert Ay\right\rVert_2
\end{align}$$

Since A is a matrix, by the definition of a matrix we have

$$\begin{align}
n &\geq 1 \\
\sqrt{n} &\geq 1
\end{align}$$

and thus can write

$$\begin{align}
\textit{sup}_{\left\lVert y\right\rVert_n = 1} \left\lVert Ay\right\rVert_{\infty} \leq \sqrt{n}\big(\textit{sup}_{\left\lVert y\right\rVert_n = 1} \left\lVert Ay\right\rVert_2\big)
\end{align}$$

Then from the definition of a matrix norm we can write

$$\begin{align}
\left\lVert A\right\rVert_{\infty} \leq \sqrt{n}\left\lVert A\right\rVert_2
\end{align}$$

**Equality condition**

For the matrix $\begin{bmatrix}a & a & \dots & a \\ 0 & 0 & \dots & 0 \\ \vdots & \vdots & \vdots & \vdots \\ 0 & 0 & \dots & 0\end{bmatrix}, a \in \mathbb{R}$, the given inequality becomes an equality.

### (d) $\left\lVert A\right\rVert_2 \leq \sqrt{m}\left\lVert A\right\rVert_{\infty}$

From **(b)** we know that for a given $m \times 1$ vector $x$, 

$$\begin{align}
\left\lVert x\right\rVert_2 \leq \sqrt{m}\left\lVert x\right\rVert_{\infty}
\end{align}$$

Let $y$ be any $n \times 1$ vector and $A$ an $m \times n$ matrix. Then $Ay$ is an $m \times 1$ vector and thus satisfies 

$$\begin{align}
\left\lVert Ay\right\rVert_2 \leq \sqrt{m}\left\lVert Ay\right\rVert_{\infty}
\end{align}$$

If we take the supremum of both sides

$$\begin{align}
\textit{sup}_{\left\lVert y\right\rVert_n = 1} \left\lVert Ay\right\rVert_2 \leq \textit{sup}_{\left\lVert y\right\rVert_n = 1} \sqrt{m}\left\lVert Ay\right\rVert_{\infty}
\end{align}$$

Since $\sqrt{m} \geq 1$, taking $\sqrt{m}$ out of the supremum doesn't change the result of the supremum function.

$$\begin{align}
\textit{sup}_{\left\lVert y\right\rVert_n = 1} \left\lVert Ay\right\rVert_2 \leq \sqrt{m}\big(\textit{sup}_{\left\lVert y\right\rVert_n = 1} \left\lVert Ay\right\rVert_{\infty}\big)
\end{align}$$

From the definition of a matrix norm we write

$$\begin{align}
\left\lVert A\right\rVert_2 \leq \sqrt{m}\left\lVert A\right\rVert_{\infty}
\end{align}$$

**Equality condition**

For the matrix $\begin{bmatrix}a & 0 & \dots & 0 \\ a & 0 & \dots & 0 \\ \vdots & \vdots & \vdots & \vdots \\ a & 0 & \dots & 0\end{bmatrix}, a \in \mathbb{R}$, the given inequality becomes an equality.
